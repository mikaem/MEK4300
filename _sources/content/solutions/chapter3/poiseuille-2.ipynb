## Solutions written assignments Poiseuille

1. Consider [problem 1](../../chapter3/poiseuille.ipynb#problem-3-15). Since the boundary conditions are independent of $x$ we may assume a parallel shear solution of the form $\boldsymbol{u} = (u(y), 0, 0)$, with $\partial p / \partial x = 0$. The momentum equation reduces to

   $$
 0 =  \mu \nabla^2u + \rho g \sin \theta,
   $$

   where $\theta$ is the angle of the plate with the horizontal. The boundary conditions are $u(0)=0$ and $du/dx(h)=0$. Integrate twice to obtain

   $$
  u = -\frac{\rho g \sin \theta y^2}{2 \mu} + Ay + B
   $$

   The boundary conditions are then used to find A and B. The first, $u(0)=0$, gives us $B=0$ and the second

   $$
\begin{aligned}
  \frac{\mathrm{d} u}{dy}(y=h) = 0 &= -\frac{\rho g \sin \theta h}{\mu} +A \\
  A &= \frac{\rho g \sin \theta h}{\mu}
\end{aligned}
   $$

   and thus the final velocity profile

   $$
  u(y) = \frac{\rho g \sin \theta y}{2 \mu}\left(2h - y\right)
   $$

   The volumetric flow rate $Q$ per unit width is found through

   $$
\begin{aligned}
  Q &= \int_0^h u(y) \mathrm{d}y \\
  Q &= \frac{\rho g h^3 \sin \theta}{3 \mu}
\end{aligned}
   $$

2. Consider [problem 2](../../chapter3/poiseuille.ipynb#problem-3-16). Assume negligible pressure gradient in $z$-direction since the film is thin and the pressure outside the film is constant atmospheric pressure. The momentum equation in $z$-direction reads

   $$
 \frac{\mu}{r}\frac{\mathrm{d} }{\mathrm{d}r}\left( r \frac{\mathrm{d} u}{\mathrm{d} r} \right) + \rho g = 0.
   $$

   Integrate twice to obtain

   $$
 u(r) = -\frac{\rho g r^2}{4 \mu} + A \ln r + B.
   $$

   The boundary conditions are no-slip at the wall and no shear at the free surface. The second one gives us

   $$
\begin{aligned}
 \frac{\mathrm{d} u}{\mathrm{d} r} (y=b) &= -\frac{\rho g b}{2 \mu} + \frac{A}{b} = 0 \\
  A &= \frac{\rho g b^2}{2 \mu}
\end{aligned}
   $$

   The no-slip boundary condition gives us

   $$
\begin{aligned}
 u(a) = 0 &= -\frac{\rho g a^2}{4 \mu} + \frac{\rho g b^2}{2 \mu} \ln a + B\\
 B &= \frac{\rho g a^2}{4 \mu} - \frac{\rho g b^2}{2 \mu} \ln a,
\end{aligned}
   $$

   which leads to the final expression for the velocity

   $$
 u(r) = \frac{\rho g}{4 \mu}\left(a^2 - r^2 \right) + \frac{\rho g b^2}{2 \mu} \ln \frac{r}{a}
   $$

   The volumetric flow rate can be found through

   $$
\begin{aligned}
  Q &= 2 \pi \int_a^b u(r) r \mathrm{d}r \\
  Q &= \ldots
\end{aligned}
   $$

3. Consider [problem 3](../../chapter3/poiseuille.ipynb#problem-3-17). Like in [problem 1](../../chapter3/poiseuille.ipynb#problem-3-15) the velocity is given as

   $$
\begin{aligned}
  u_1 &= -\frac{\rho_1 g \sin \theta y^2}{2 \mu_1} + A_1y + B_1, \\
  u_2 &= -\frac{\rho_2 g \sin \theta y^2}{2 \mu_2} + A_2y + B_2.
\end{aligned}
   $$

   The boundary conditions are now

   $$
\begin{aligned}
 u_1(0) &= 0, \\
 u_1(h_1) &= u_2(h_1), \\
 \mu_1 \frac{\mathrm{d} u_1}{\mathrm{d} y}(h_1) &= \mu_2 \frac{\mathrm{d} u_2}{\mathrm{d} y}(h_1) \\
 \frac{\mathrm{d} u_2}{\mathrm{d} y}(h_1+h_2) &= 0.
\end{aligned}
   $$

   The first one gives $B_1=0$. The remaining leads to three equations for three unknowns. They are:

   $$
\begin{aligned}
 -\frac{\rho_1 g \sin \theta h_1^2}{2 \mu_1} + A_1h_1 &= -\frac{\rho_2 g \sin \theta h_1^2}{2 \mu_2} + A_2h_1 + B_2, \\
 -\frac{\rho_1 g \sin \theta h_1}{\mu_1} +A_1 &= -\frac{\rho_2 g \sin \theta h_1}{\mu_2} +A_2 \\
 -\frac{\rho_2 g \sin \theta (h_1+h_2)}{\mu_2} + A_2 &= 0
\end{aligned}
   $$

   where $A_2$ is given by the last equation. Inserting this into the previous gives us $A_1$

   $$
\begin{aligned}
  A_1 &= -\frac{\rho_2 g \sin \theta h_1}{\mu_2} + \frac{\rho_2 g \sin \theta (h_1+h_2)}{\mu_2} + \frac{\rho_1 g \sin \theta h_1}{\mu_1}\\
 A_1 &= g \sin \theta \left(\frac{\rho_2 h_2}{\mu_2} + \frac{\rho_1 h_1}{\mu_1} \right)
\end{aligned}
   $$

   And $A_1$ and $A_2$ inserted into the first gives us $B_2$

   $$
 B_2 = -\frac{\rho_1 g \sin \theta h_1^2}{2 \mu_1} + A_1 h_1 + \frac{\rho_2 g \sin \theta h_1^2}{2 \mu_2} - A_2 h_1
   $$


## Solutions computer exercises Poiseuille

1. Verify analytical solutions ([computer problem 1](../../chapter3/poiseuille.ipynb#p-computer-1))

   In the script below we compute the solution for the triangle case, i.e., we verify Eq. (3-49) in {cite}`white06`. We create a script that computes the L2 error norm for a given mesh refinement and order of approximating polynomials.

```python
import sys
import os
from dolfin import *
from mshr import *
from numpy import cosh , cos, sin, radians, sqrt, log
import matplotlib.pyplot as plt
import numpy as np

plotting = True

def poisson_triangle(refinement, order, mesh_density=1):
    print("Refinement level = ", refinement)
    print("Order = ", order)

    dpdx = Constant(-0.01)
    mu = Constant(0.01)
    a = 1.
    h = (a/2)*sqrt(3.)

    p0 = Point(0, 0)
    p1 = Point(-a/2, -h)
    p2 = Point(a/2, -h)
    domain = Polygon([p0, p1, p2, p0])
    mesh = generate_mesh(domain, mesh_density)
    for i in range(refinement):
        mesh = refine(mesh)
    
    if plotting: 
        plt.figure(figsize=(12, 4))
        plt.subplot(131)
        plot(mesh, title='Mesh')

    V = FunctionSpace(mesh, 'CG', order)
    u = TrialFunction(V)
    v = TestFunction(V)

    F = inner(grad(u), grad(v))*dx + 1/mu*dpdx*v*dx

    bc = DirichletBC(V, Constant(0), DomainBoundary())

    u_ = Function(V)
    solve(lhs(F) == rhs(F), u_, bcs=bc)

    if plotting:
        plt.subplot(132)
        ax1 = plot(u_, title="Velocity normal to plane")

    u_exact = Expression("dpdx/2./sqrt(3.)/a/mu*(x[1]+0.5*a*sqrt(3.))*(3*x[0]*x[0]-x[1]*x[1])",
                         mu=mu(0), dpdx=dpdx(0), a=a, degree=3)

    u_e = interpolate(u_exact, V)
    if plotting:
        plt.subplot(133)
        ax2 = plot(u_e-u_, title="Error velocity")
        plt.colorbar(ax2, fraction=0.045)

    u_error = errornorm(u_, u_e, degree_rise=1)
    #u_error = np.linalg.norm(u_.vector().get_local()-u_e.vector().get_local())**2

    Q_comp  = assemble(u_*dx)
    Q_exact = a**4*sqrt(3)/320./mu*(-dpdx)

    print("Computed volume flow = ", Q_comp)
    print("Exact volume flow    = ", Q_exact(0))
    print("Computed error = %2.8e %2.8e " %(u_error, mesh.hmin()))

    if plotting: plt.show()
    
    return u_error, mesh.hmin()
```

We can now call `poisson_triangle` with three arguments, the refinement level, the approximation order, and finally a mesh density $(>0)$ to start the refinement from. 

```python
mesh_density = 8
refinement = 1
order = 1
u_error, h = poisson_triangle(refinement, order, mesh_density)
```

We see that the mesh is inhomogeneous, i.e., the cells are not all the same size. Let's try again with a different value for the `mesh_density` 

```python
u_error, h = poisson_triangle(refinement, order, 1)
```

Wow! What happened here? We seem to approximate the solution exactly, or at least up to machine precision. If you look at the exact solution, you'll see that it's a third order polynomial. But we used piecewise linear elements, so how can the numerical solution be exact? An important observation is that the mesh now is uniform, i.e.,  all elements have the same size. And what we observe is actually something called superconvergence, that only happens for uniform meshes. If we make the mesh non-uniform, then we get what is expected and we need to use third order elements to achive machine precision, because the solution is a third order polynomial. Change the order to 3 and mesh_density to 4 to see the expected behaviour.

For a nonuniform mesh we can investigate the order of accuracy using the technique discussed in [Analysing the error](../../chapter3/poiseuille.ipynb#analysing-the-error).

```python
hs, errors = [], []
# Suppress plotting and printing
plotting = False
stdout = sys.stdout
f = open('/dev/null', 'w')
sys.stdout = f
for refinement in (1, 2, 3, 4):
    u_e, hmin = poisson_triangle(refinement, 1, 4)
    errors.append(u_e)
    hs.append(hmin)

# Enable printing again
sys.stdout = stdout
for i, (error, h) in enumerate(zip(errors, hs)):
    if i == 0:
        print("Error          hmin           r       ")
        print("%2.8e %2.8e %2.8f"%(error, h, 0))
    else:
        print("%2.8e %2.8e %2.8f"%(error, h, log(error/errors[i-1])/log(h/hs[i-1])))
```
